```python
from player_data import player_rankings
from match_data import matches


players = list(player_rankings.keys())
top_players = [player.split(",")[1].lower().strip() + " " + player.split(",")[0].lower().strip()  for player in players]
top_players += [player.split(",")[0].lower().strip() + " " + player.split(",")[1].lower().strip()  for player in players]

useful_matches = []
for match in matches:

    if match[0].lower() in top_players and match[1].lower() in top_players :
         useful_matches.append(match)


```

```python
for player in list(player_rankings.keys()):
    player_rankings[player.split(",")[1].lower().strip() + " " + player.split(",")[0].lower().strip()] = player_rankings[player]
    player_rankings[player.split(",")[0].lower().strip() + " " + player.split(",")[1].lower().strip()] = player_rankings[player]

```

```python
ultimate_data = useful_matches.copy()
```

```python
for match in ultimate_data:
    name1 = match[0]
    name2 = match[1]

    temp1 = player_rankings[match[0].lower()][1:]
    temp2 = player_rankings[match[1].lower()][1:]
    
    match[0] = temp1
    match[1] = temp2
    
ultimate_data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(ultimate_data, columns=["player 1", "player 2", "match score", "winner"])

df.dropna()
```

```python
df.drop(axis=1, columns="winner", inplace=True)
```

```python
def winner(x):
    scores = x.split(":")
    try:
        if scores[0] > scores[1]:
            return 1
        else:
            return 0
    except:
        return 0

df["winner"] = df["match score"].apply(winner)
```

```python
def p1_score(x):
    try:
        scores = x.split(":")
        return scores[0]
    except:
        return 0
df["p1_score"] = df["match score"].apply(p1_score)

def p2_score(x):
    try:
        scores = x.split(":")
        return scores[1]
    except:
        return 0
df["p2_score"] = df["match score"].apply(p2_score)


df
```

```python
p1_df = df["player 1"]
p2_df = df["player 2"]
```

```python
def ranking(x):
    return x[0]

def country(x):
    return x[1]

def age(x):
    return x[2]

p1_data = pd.DataFrame()

p1_data["ranking"] = p1_df.apply(ranking)
p1_data["country"] = p1_df.apply(country)
p1_data["age"] = p1_df.apply(age)

```

```python
p1_data = p1_data.drop(["ranking", "country", "age"])
```

```python
def ranking(x):
    return x[0]

def country(x):
    return x[1]

def age(x):
    return x[2]

p2_data = pd.DataFrame()

p2_data["ranking"] = p2_df.apply(ranking)
p2_data["country"] = p2_df.apply(country)
p2_data["age"] = p2_df.apply(age)
```

```python
p1_data = pd.get_dummies(data=p1_data, columns=["country"])
```

```python
p2_data = pd.get_dummies(data=p2_data, columns=["country"])
```

```python
p2_data
```

```python
final_data = pd.concat([p1_data, p2_data],axis=1)
final_data.dropna(axis=1, inplace=True)
```

```python
results = df["winner"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(final_data, results, test_size=0.33)
```

```python
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression
```

```python
clf_A = LogisticRegression()
```

```python
clf_A.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
predictions = clf_A.predict(X_test)
```

```python
print(classification_report(y_test, predictions))
```
